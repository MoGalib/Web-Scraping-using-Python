# Web Scraping in PYTHON using Requests module ###

```python
#import requests module
import requests
```

```python
url = "https://www.manutd.com/"
```

```python
#Decide on the word we want to scrape from the website
word = "Manchester United"
```

```python
response = requests.get(url)
```

```python
#print the response status code to see if our url is functioning properly
print(response.status_code)
```

##### response is okay 

```python
#print the whole url body and tags
t = response.text
print(t)
```

```python
#Finding the index of the word in the website
ind = t.index(word)
print(ind)
```

```python
rt = t[ind:]
print(rt)
```

```python
#Creating a list by splitting the text 
rt = t[ind:].split("Website")
print(rt)
```

```python
rt = t[ind:].split("Website")[0]
print(rt)
```

## We have scraped through the website to find the word we are looking for successfully.
